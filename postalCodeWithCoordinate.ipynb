```python
import pandas as pd
import numpy as np
```

* First, we extract csv with Toronto geographical coordinates to dataframe
* furtermore we will initialize the postal code data we got from wikipedia
* to finish we will combine the postCode with the geographical coordinates

```python
# getting toronto geographical data
lat_long_data = 'https://cocl.us/Geospatial_data'
!wget -q -O 'toronto_m.geospatial_data.csv' lat_long_data
coordinate_data = pd.read_csv(lat_long_data).set_index("Postal Code")
coordinate_data.head()
```

```python
# initialization of the postal data got from wikipedia
PostalCodes = pd.read_csv('Toronto_Postal_data.csv').set_index("PostalCode")
PostalCodes.rename_axis("Postal Code", axis='index', inplace=True)
PostalCodes.head()
```

```python
PostalCodesWithCoordinates = PostalCodes.join(coordinate_data)
PostalCodesWithCoordinates.head(10)
```

```python
PostalCodesWithCoordinates.to_csv('PostalCodesWithCoordinates.csv', index=False)
```
